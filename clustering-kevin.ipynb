```python
import os
import json
import random

incidentList = json.loads(open('data/vcdb.json').read())

random.seed("follow @bfist on twitter")
```

```python
enum = json.loads(open("veris/verisc-enum.json").read())
```

```python
def normalizeNAICS(inString):
    """The naics codes can be anywhere from two digits to six digits and can also include three
    special ranges. This will return just the first two characters unless a range is specified"""
    if "-" in inString:
        return inString
    else:
        return inString[0:2]
```

```python
def flatVERIS(obj):
    output = {}
    # randoms
    for eachRandom in ['confidence', 'cost_corrective_action', 'discovery_method', 'security_incident', 'targeted']:
        for eachEnum in enum[eachRandom]:
            if obj.get(eachRandom,"") == eachEnum:
                output['.'.join([eachRandom,eachEnum])] = 1
    
    # all the action properties
    for eachAction in ['hacking', 'malware', 'social', 'error', 'misuse', 'physical','environmental']:
        if eachAction in obj['action'].keys():
            output['action.'+eachAction] = 1
        for eachEnum in enum['action'][eachAction].get('variety', []):
            if eachEnum in obj['action'].get(eachAction, {}).get('variety', []):
                output['.'.join(['action',eachAction,'variety',eachEnum])] = 1
        for eachEnum in enum['action'][eachAction].get('vector', []):
            if eachEnum in obj['action'].get(eachAction, {}).get('vector', []):
                output['.'.join(['action',eachAction,'vector',eachEnum])] = 1
        if eachAction == "social":
            for eachEnum in enum['action']['social']['target']:
                if eachEnum in obj['action'].get('social', {}).get('target', []):
                    output['.'.join(['action',eachAction,'target',eachEnum])] = 1
    
    # now the actor properties
    for eachActor in ['external','internal','partner']:
        for eachEnum in enum['actor']['motive']:
            if eachEnum in obj['actor'].get(eachActor, {}).get('motive', []):
                output['.'.join(['actor',eachActor,'motive',eachEnum])] = 1
        for eachEnum in enum['actor'].get(eachActor, {}).get('variety', []):
            if eachEnum in obj['actor'].get(eachActor, {}).get('variety', []):
                output['.'.join(['actor',eachActor,'variety',eachEnum])] = 1
        for eachEnum in enum['country']:
            if eachEnum in obj['actor'].get(eachActor, {}).get('country', {}):
                output['.'.join(['actor',eachActor,'country',eachEnum])] = 1
        if eachActor == 'partner':
            if len(obj['actor'].get('partner', {}).get('industry','')) > 0:
                   shortNAICS = normalizeNAICS(obj['actor']['partner']['industry'])
                   output['actor.partner.industry.'+shortNAICS] = 1
            for each in obj['actor'].get('partner', {}).get('region', []):
                output['actor.partner.region.' + each[0:3]] = 1
    
    # now for the attribute properties
    for eachEnum in enum['attribute']['availability']['variety']:
        if eachEnum in obj['attribute'].get('availability', {}).get('variety', []):
            output['attribute.availability.variety.'+eachEnum] = 1
    for eachEnum in enum['timeline']['unit']:
        if eachEnum == obj['attribute'].get('availability', {}).get('duration', {}).get('unit',''):
            output['attribute.availability.duration.unit.'+eachEnum] = 1
    for eachObject in obj['attribute'].get('confidentiality', {}).get('data', []):
        output['attribute.confidentiality.data.variety.'+eachObject['variety']] = 1
    if obj['attribute'].get('confidentiality',{}).get('data_disclosure', '') != '':
        output['attribute.confidentiality.data_disclosure.'+obj['attribute']['confidentiality']['data_disclosure']] = 1
    for eachEnum in enum['attribute']['confidentiality']['data_victim']:
        if eachEnum in obj['attribute'].get('confidentiality', {}).get('data_victim',[]):
            output['attribute.confidentiality.data_victim.'+eachEnum] = 1
    for eachEnum in enum['attribute']['confidentiality']['state']:
        if eachEnum in obj['attribute'].get('confidentiality', {}).get('state', []):
            output['attribute.confidentiality.state.'+eachEnum] = 1
    for eachEnum in enum['attribute']['integrity']['variety']:
        if eachEnum in obj['attribute'].get('integrity', {}).get('variety', []):
            output['attribute.integrity.variety.'+eachEnum] = 1
    
    # asset properties
    for eachAsset in obj['asset'].get('assets', []):
        for eachEnum in enum['asset']['variety']:
            if eachAsset.get('variety','') == eachEnum:
                output['asset.assets.variety.'+eachEnum] = 1
    for eachEnum in enum['asset']['cloud']:
        if eachEnum == obj['asset'].get('cloud',''):
            output['asset.cloud.'+eachEnum] = 1
    for eachEnum in enum['asset']['governance']:
        if eachEnum == obj['asset'].get('governance',''):
            output['asset.governance.'+eachEnum] = 1
    for eachEnum in enum['country']:
        if eachEnum in obj['asset'].get('country',[]):
            output['asset.country.'+eachEnum] = 1
            
    # impact properties
    for eachEnum in enum['iso_currency_code']:
        if eachEnum == obj.get('impact',{}).get('iso_currency_code',''):
            output['impact.iso_currency_code.'+eachEnum] = 1
    for eachEnum in enum['impact']['overall_rating']:
        if eachEnum == obj.get('impact', {}).get('overall_rating',''):
            output['impact.overall_rating.'+eachEnum] = 1
    for eachLoss in obj.get('impact', {}).get('loss', []):
        output['impact.loss.variety.'+eachLoss['variety']] = 1
    
    # timeline properties
    for each in ['compromise', 'containment', 'discovery', 'exfiltration']:
        if obj['timeline'].get(each, {}).get('unit','') != '':
            output['.'.join(['timeline',each,'unit',obj['timeline'][each]['unit']])] = 1
    output['timeline.incident.year.'+str(obj['timeline']['incident']['year'])] = 1

    # victim data
    for each in obj['victim'].get('country',[]):
        output['victim.country.'+each] = 1
    for each in obj['victim'].get('region', []):
        output['victim.region.'+each[0:3]] = 1
    shortNAICS = normalizeNAICS(obj['victim'].get('industry','000'))
    output['victim.industry.'+shortNAICS] = 1
    if obj['victim'].get('employee_count', '') != '':
        output['victim.employee_count.' + obj['victim']['employee_count']] = 1
            
    return output
```

```python
for index, each in enumerate(incidentList):
    flatVERIS(each)
```

```python
df = DataFrame([flatVERIS(i) for i in incidentList])
df = df.fillna(0)
```

```python
# with all the rows this step took about 1:30 on my machine
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
from sklearn import manifold
euc = euclidean_distances(df)
man = manhattan_distances(df)
```

```python
# mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, 
#       dissimilarity="precomputed", n_jobs=1)
mds = manifold.MDS()
```

```python
# 500 rows took 36 seconds
# 600 rows took 1 minute 14 seconds
# 700 rows also took 1 minute 14 seconds. Weird.
# 1000 rows took 3 minutes 17 seconds.
# 2000 rows took at least 15:00 and ended before 17 minutes.
pos_e = mds.fit(euc).embedding_
pos_m = mds.fit(man).embedding_
```

```python
df['xpos_e'] = pos_e[:,0]
df['ypos_e'] = pos_e[:,1]
df['xpos_m'] = pos_m[:,0]
df['ypos_m'] = pos_m[:,1]
```

```python
def identifyPattern(inDict):
    # skimmers
    if inDict["attribute.confidentiality.data.variety.Payment"] == 1 and \
      inDict["action.physical.variety.Tampering"]==1:
            return '#283F6C'  # dark blue
    
    # dos
    if inDict['action.hacking.variety.DoS'] == 1:
        return '#CC548D' # pink
    
    if inDict["action.hacking.vector.Web application"]==1:
        return '#E37B4F'
    
    # loss and theft
    if inDict['action.error.variety.Loss'] == 1 or inDict['action.physical.variety.Theft'] == 1:
        return '#45B2A1'
    
    # espionage
    if inDict['actor.external.motive.Espionage'] == 1 or \
      inDict["actor.external.variety.State-affiliated"]==1 :
        return '#74406D' # purple
    
    # crimeware
    if inDict['action.malware'] == 1 and \
      inDict['action.malware.vector.Direct install'] == 0:
            return '#45B4F7' # light blue
    
    # misuse
    if inDict['action.misuse'] == 1:
        return '#FAA85C' # yellow
    
    # point of sale
    if inDict['action.hacking'] == 1 and \
      (inDict["asset.assets.variety.S - POS controller"]==1 or inDict["asset.assets.variety.U - POS terminal"]==1):
            return '#8A231A' # red
    
    # Error
    if inDict['action.error'] == 1:
        return '#2B9A5B'
    
    # Other
    return 'white' # nothing else will show up because of white on white
```

```python
df['pattern'] = [identifyPattern(row) for index, row in df.iterrows()]
```
