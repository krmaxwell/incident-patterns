```python
def normalizeNAICS(inString):
    """The naics codes can be anywhere from two digits to six digits and can also include three
    special ranges. This will return just the first two characters unless a range is specified"""
    if "-" in inString:
        return inString
    else:
        return inString[0:2]
```

```python
def flatVERIS(obj,exclude_unknown=True,exclude_country=True,exclude_timeline=True):
    output = {}
    ignore = ['']
    if exclude_unknown:
        ignore.append('Unknown')
    # randoms
    for eachRandom in ['confidence', 'cost_corrective_action', 'discovery_method', 'security_incident', 'targeted']:
        if obj.get(eachRandom,'Unknown') not in ignore:
            output['.'.join([eachRandom,obj[eachRandom]])] = 1
    
    # all the action properties
    for eachAction in ['hacking', 'malware', 'social', 'error', 'misuse', 'physical','environmental']:
        if eachAction in obj['action'].keys():
            output['action.'+eachAction] = 1
        for eachSubVariable in ['variety','vector','target']:
            for eachEnum in obj['action'].get(eachAction, {}).get(eachSubVariable, []):
                if eachEnum not in ignore:
                    output['.'.join(['action',eachAction,eachSubVariable,eachEnum])] = 1

    
    # now the actor properties
    for eachActor in ['external','internal','partner']:
        if eachActor in obj['actor'].keys():
            output['.'.join(['actor',eachActor])] = 1
        for eachSubVariable in ['motive','variety','country']:
            if eachSubVariable == 'country' and exclude_country:
                continue
            for eachEnum in obj['actor'].get(eachActor, {}).get(eachSubVariable, []):
                if eachEnum not in ignore:
                    output['.'.join(['actor',eachActor,eachSubVariable,eachEnum])] = 1
        if eachActor == 'partner':
            if len(obj['actor'].get('partner', {}).get('industry','')) > 0:
                   shortNAICS = normalizeNAICS(obj['actor']['partner']['industry'])
                   output['actor.partner.industry.'+shortNAICS] = 1
            for each in obj['actor'].get('partner', {}).get('region', []):
                output['actor.partner.region.' + each[0:3]] = 1
    
    # now for the attribute properties
    for eachAttribute in ['confidentiality','integrity','availability']:
        for eachSubVariable in ['variety','state','data_victim']:
            for eachEnum in obj['attribute'].get(eachAttribute, {}).get(eachSubVariable, []):
                if eachEnum not in ignore:
                    output['.'.join(['attribute',eachAttribute,eachSubVariable,eachEnum])] = 1
    if eachAttribute == 'availability':
        unit = obj['attribute'].get('availaiblity',{}).get('duration',{}).get('unit','')
        if unit not in ignore:
            output['attribute.availability.duration.unit.'+unit] = 1
    for eachObject in obj['attribute'].get('confidentiality', {}).get('data', []):
        if eachObject.get('variety','Unknown') not in ignore:
            output['attribute.confidentiality.data.variety.'+eachObject['variety']] = 1
    if obj['attribute'].get('confidentiality',{}).get('data_disclosure','') not in ignore:
        value = obj['attribute'].get('confidentiality',{}).get('data_disclosure','')
        output['attribute.confidentiality.data_disclosure.'+value] = 1
   
    # asset properties
    for eachAsset in obj['asset'].get('assets', []):
        if eachAsset['variety'] not in ignore:
            output['asset.assets.variety.' + eachAsset['variety']] = 1
    if obj['asset'].get('cloud','') not in ignore:
        value = obj['asset']['cloud']
        output['asset.cloud.'+value] = 1
    for eachVariable in ['country','governance']:
        for eachItem in obj['asset'].get(eachVariable, []):
            if eachItem not in ignore:
                output['.'.join(['asset',eachVariable,eachItem])] = 1
         
    # impact properties
    for eachVariable in ['iso_currency_code','overall_rating']:
        if obj.get('impact',{}).get(eachVariable,'Unknown') not in ignore:
            output['.'.join(['impact',eachVariable,obj['impact'][eachVariable]])] = 1
    for eachLoss in obj.get('impact', {}).get('loss', []):
        output['impact.loss.variety.'+eachLoss['variety']] = 1
    
    # timeline properties
    if not exclude_timeline:
        for each in ['compromise', 'containment', 'discovery', 'exfiltration']:
            if obj['timeline'].get(each, {}).get('unit','') not in ignore:
                output['.'.join(['timeline',each,'unit',obj['timeline'][each]['unit']])] = 1
        output['timeline.incident.year.'+str(obj['timeline']['incident']['year'])] = 1

    # victim data
    if not exclude_country:
        for each in obj['victim'].get('country',[]):
            if each not in ignore:
                output['victim.country.'+each] = 1
    for each in obj['victim'].get('region', []):
        output['victim.region.'+each[0:3]] = 1
    shortNAICS = normalizeNAICS(obj['victim'].get('industry','000'))
    output['victim.industry.'+shortNAICS] = 1
    if obj['victim'].get('employee_count', '') != '':
        output['victim.employee_count.' + obj['victim']['employee_count']] = 1
            
    return output
```

```python
for index, each in enumerate(incidentList):
    flatVERIS(each)
```

```python
from pandas import *
```

```python
df = DataFrame([flatVERIS(i) for i in incidentList])
df = df.fillna(0)
```

```python
# with all the rows this step took about 1:30 on my machine
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
from sklearn import manifold
euc = euclidean_distances(df)
man = manhattan_distances(df)
```

```python
# mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, 
#       dissimilarity="precomputed", n_jobs=1)
mds = manifold.MDS()
```

```python
# 500 rows took 36 seconds
# 600 rows took 1 minute 14 seconds
# 700 rows also took 1 minute 14 seconds. Weird.
# 1000 rows took 3 minutes 17 seconds.
# 2000 rows took at least 15:00 and ended before 17 minutes.
pos_e = mds.fit(euc).embedding_
pos_m = mds.fit(man).embedding_
```

```python
df['xpos_e'] = pos_e[:,0]
df['ypos_e'] = pos_e[:,1]
df['xpos_m'] = pos_m[:,0]
df['ypos_m'] = pos_m[:,1]
```

```python
def identifyPattern(inDict):
    # skimmers
    if inDict["attribute.confidentiality.data.variety.Payment"] == 1 and \
      inDict["action.physical.variety.Tampering"]==1:
            return 'skimmer'  # dark blue
    
    # dos
    if inDict['action.hacking.variety.DoS'] == 1:
        return 'dos' # pink
    
    if inDict["action.hacking.vector.Web application"]==1:
        return 'webapp'
    
    # loss and theft
    if inDict['action.error.variety.Loss'] == 1 or inDict['action.physical.variety.Theft'] == 1:
        return 'losstheft'
    
    # espionage
    if inDict['actor.external.motive.Espionage'] == 1 or \
      inDict["actor.external.variety.State-affiliated"]==1 :
        return 'espionage' # purple
    
    # crimeware
    if inDict['action.malware'] == 1 and \
      inDict['action.malware.vector.Direct install'] == 0:
            return 'crimeware' # light blue
    
    # misuse
    if inDict['action.misuse'] == 1:
        return 'misuse' # yellow
    
    # point of sale
    if inDict['action.hacking'] == 1 and \
      (inDict["asset.assets.variety.S - POS controller"]==1 or inDict["asset.assets.variety.U - POS terminal"]==1):
            return 'pos' # red
    
    # Error
    if inDict['action.error'] == 1:
        return 'error'
    
    # Other
    return 'other' # nothing else will show up because of white on white
```

```python
df['pattern'] = [identifyPattern(row) for index, row in df.iterrows()]
```

This is the multi-dimensional scaling using the "manhattan" method that Jay uses in his blog post

```python
from matplotlib import pyplot as plt
%matplotlib inline
fig, ax = ppl.subplots(figsize=(10,6))
for each in ['skimmer','dos','webapp','losstheft','espionage','crimeware','misuse','pos']:
    ypos = df[df['pattern']==each]['ypos_m']
    xpos = df[df['pattern']==each]['xpos_m']
    ax = ppl.scatter(ypos,xpos,label=each)
ppl.legend(ax,loc='upper right')
plt.title("Multi-dimensional Scaling: Manhattan")
plt.show()
```
