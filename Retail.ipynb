# Retail organizations

We analyzed the [VERIS Community Database](http://vcdb.org) for several examples of common incident patterns. This notebook dives more deeply into incidents in retail organizations.

Again, we load in the full set of VCDB incidents. We'll also load some useful modules for data plotting and define some useful functions.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
makePretty = True # if you don't have prettyplotlib installed then make this False
```

```python
def plot(data, title):
    if makePretty:
        import prettyplotlib as ppl
    labels = sorted(data, key=data.get)
    values = [data[x] for x in labels]
    ypos = np.arange(len(data))+.1
    fig, ax = plt.subplots(1)
    if makePretty:
        ppl.barh(ax, ypos, values)
    else:
        plt.barh(ypos,values,color="red",align="center",height=.9)
    
    # Adds some padding to the bars so that it doesn't butt up 
    # against the edge of the image
    plt.xlim(0,max(values)*1.1)
    plt.yticks(ypos,labels)
    plt.title(title)
    plt.show()
```

```python
import json
with open('data/vcdb.json','rb') as f:
    vcdb = json.load(f)
print "%d incidents imported" % len(vcdb)
```

The NAICS codes for retail organizations are `44` and `45`.

```python
retail_attacks = []
for each in vcdb:
    industry = each['victim']['industry'][:2]
    if industry == '44' or industry == '45':
        retail_attacks.append(each)
print "%d incidents in the retail industry" % len(retail_attacks)
```

## Vectors and Varieties

Let's do a quick overview of the actions.

```python
from collections import Counter

retail_actions = {}
retail_actions['varieties'] = Counter()
retail_actions['vectors'] = Counter()
for attack in retail_attacks:
    for action in attack['action']:
        for variety in attack['action'][action].get('variety', {}):
            retail_actions['varieties']['%s.%s' % (action,variety)] += 1
        for vector in attack['action'][action].get('vector', {}):
            retail_actions['vectors']['%s.%s' % (action,vector)] += 1
print retail_actions['vectors']
print '---'
print retail_actions['varieties']
```

The **Unknown** category is higher than we'd like, but those are the data we have at hand. Let's do a quick chart to get a better understanding.

```python
plot(retail_actions['vectors'], 'Vectors in retail attacks. n = %d' % len(retail_attacks))
```

```python
print "%d unknown varieties of hacking" % retail_actions['varieties']['hacking.Unknown']
print "%d unknown varieties of malware"  % retail_actions['varieties']['malware.Unknown']
del retail_actions['varieties']['hacking.Unknown']
del retail_actions['varieties']['malware.Unknown']
plot(retail_actions['varieties'], 'Varieties in retail attacks. n = %d' % len(retail_attacks))
```

```python
for each in retail_attacks:
    if 'Unknown' in each['action'].get('hacking',{}).get('variety',{}):
        print each['action']['hacking']
```

## Discovery Methods

```python
methods = Counter()
for incident in retail_attacks:
    methods[incident['discovery_method']] += 1

print "%d incidents discovered by unknown methods" % methods['Unknown']
del methods['Unknown']
plot(methods, "Discovery Methods")
```

Generally, customer-disclosed incidents relate to financially-motivated actors (identity theft and fraud), while actor-disclosed incidents often relate to national security disclosures (e.g. the Manning and Snowden cases) or "fun" that gets disclosed on social media.
